# Class to Cardio: Letting ChatGPT Coach My Weekly Workouts



## Introduction
As a full time college student, I don’t have hours to craft the perfect routine, but I still want to keep moving and stay at my current weight. Could Chat step in as a virtual trainer and hand me a balanced, college‑friendly program that fits a busy campus schedule?

---

## Data Used  

- **Stats:** 21 y/o female, 5' 7.5", 125 lbs, college student  
- **Goals:** Maintain current weight, support overall health, boost basic strength and cardio  
- **Constraints:**  
  * 4 workout days/week  
  * Access to standard commercial gym gear (barbells, dumbbells, treadmill)  

---

## The Prompt I Gave GPT‑4o  

> “Design a 4‑week strength‑and‑cardio program for a 21‑year‑old woman (5' 7.5", 125 lbs, college student) 
> - 4 gym days/week, campus‑friendly time budget
> - Goals: maintain current weight, stay healthy, improve baseline fitness 
> - Output: weekly schedule plus session detail (sets, reps, rest)”

---

## GPT‑4o’s 4‑Week Blueprint (summarized)  

> **Week Structure**
> 
> - **Day 1 (Full‑Body A):** Goblet squat 3×10 (box height to spare knee), Dumbbell bench 3×8, Bent‑over row 3×10, Plank circuit, 15‑min brisk treadmill walk.
> 
> - **Day 2 (Cardio Intervals):** 5‑min warm‑up, Treadmill intervals 6×1‑min fast / 1‑min walk, Cool‑down stretch.
> 
> - **Day 3 (Full‑Body B):** Romanian deadlift 3×10, Overhead press 3×8, Step‑ups 3×12/leg, Lat pull‑down 3×12, Farmer carry 2×40 m.
> 
> - **Day 4 (Steady‑State & Core):** 2‑mile tempo run or 25‑min elliptical, then core finisher (hanging knee raise, side plank).  
>
> **Progression Notes**
> 
> - Add 5 lb to lifts when last rep feels ≤ RPE 7.
>   
> - Week‑by‑week mileage: 5 → 6 → 7 → 8 miles.
>   
> - Substitute step‑ups for lunges if knee pain flares.

---

## Observations  
* **Time‑efficient:** Each session clocks in under 60 minutes—doable between classes.  
* **Fancy terms:** As someone who isnt familiar with many gym objects and terms this was a little complicated to dissect.
* **Nutrition omitted:** Maintaining weight also means caloric awareness, outside the scope but worth noting.

---

## Key Takeaways  

1. **Fast personalization:** One prompt delivered a four‑week plan tuned to goals and schedule.
2. **Additional nuance needed:** No deload or nutrition advice
3. **Prompt power:** Specific constraints (“campus‑friendly length”) shaped a realistic routine.

---

## Why Does This Matter?  
For students on tight budgets and tighter timetables, an LLM can kick‑start fitness planning, saving hours of Googling cookie‑cutter routines.
