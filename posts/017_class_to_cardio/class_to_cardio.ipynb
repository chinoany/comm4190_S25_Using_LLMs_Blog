# Class to Cardio: Letting ChatGPT Coach My Weekly Workouts



## Introduction
As a full time college student, I dont have hours to craft the perfect routine, but I still want to keep moving and stay at my current weight. Could Chat step in as a virtual trainer and hand me a balanced, collegefriendly program that fits a busy campus schedule?

---

## Data Used  

- **Stats:** 21y/o female, 5' 7.5", 125lbs, college student  
- **Goals:** Maintain current weight, support overall health, boost basic strength and cardio  
- **Constraints:**  
  * 4 workout days/week  
  * Access to standard commercial gym gear (barbells, dumbbells, treadmill)  

---

## The Prompt I Gave GPT4o  

> Design a 4week strengthandcardio program for a 21yearold woman (5' 7.5", 125lbs, college student) 
> - 4 gym days/week, campusfriendly time budget
> - Goals: maintain current weight, stay healthy, improve baseline fitness 
> - Output: weekly schedule plus session detail (sets, reps, rest)

---

## GPT4os 4Week Blueprint (summarized)  

> **Week Structure**
> 
> - **Day1 (FullBody A):** Goblet squat 310 (box height to spare knee), Dumbbell bench 38, Bentover row 310, Plank circuit, 15min brisk treadmill walk.
> 
> - **Day2 (Cardio Intervals):** 5min warmup, Treadmill intervals 61min fast / 1min walk, Cooldown stretch.
> 
> - **Day3 (FullBody B):** Romanian deadlift 310, Overhead press 38, Stepups 312/leg, Lat pulldown 312, Farmer carry 240m.
> 
> - **Day4 (SteadyState & Core):** 2mile tempo run or 25min elliptical, then core finisher (hanging knee raise, side plank).  
>
> **Progression Notes**
> 
> - Add 5lb to lifts when last rep feels RPE7.
>   
> - Weekbyweek mileage: 5678 miles.
>   
> - Substitute stepups for lunges if knee pain flares.

---

## Observations  
* **Timeefficient:** Each session clocks in under 60minutesdoable between classes.  
* **Fancy terms:** As someone who isnt familiar with many gym objects and terms this was a little complicated to dissect.
* **Nutrition omitted:** Maintaining weight also means caloric awareness, outside the scope but worth noting.

---

## Key Takeaways  

1. **Fast personalization:** One prompt delivered a fourweek plan tuned to goals and schedule.
2. **Additional nuance needed:** No deload or nutrition advice
3. **Prompt power:** Specific constraints (campusfriendly length) shaped a realistic routine.

---

## Why Does This Matter?  
For students on tight budgets and tighter timetables, an LLM can kickstart fitness planning, saving hours of Googling cookiecutter routines.
