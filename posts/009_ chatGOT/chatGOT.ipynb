# ChatGOT: ChatGPT Writes an imagined Season 9 *Game of Thrones* Novella!

## Introduction
Have you ever wished an epic series never ended, craving just *one more* season after the screen fades to black? As a fan of the infamous *Game of Thrones* series I found myself in this position after the show ended abrutly and rushed in the eighth and final season. With a single prompt I asked ChatGPT to produce a mini book version of an imagined Season 9 of *Game of Thrones* and package it as a PDF. After downloading and reading the 8 chapter novella, I comment on whether it scratches that post-finale itch!

---

## Experiment Setup
- **Model:** GPT-4o  
- **Prompt:** “Write a short novella (~1,700 words) that tells a complete Season 9 story of *Game of Thrones*, broken into chapters.” Format output for PDF, no pictures.

---


## You get to be one of the first to read the Game of Thrones Season 9 mini book!

```python
from IPython.display import IFrame, display
import base64, pathlib

pdf_path = pathlib.Path("GoT_Season9_Novella.pdf")

data = base64.b64encode(pdf_path.read_bytes()).decode()

display(IFrame(f"data:application/pdf;base64,{data}", width=900, height=550))
```

---

## Observations After Reading
1. **Lore continuity is spot on.** The novella picks up right after Bran’s coronation, weaves in politics (Northern autonomy, Iron Bank debts) and magic (frost spiders) all true to how Season 8 left off.  
2. **Character voices feel authentic.**
3. **Not as detailed as I would like.** At ~1,700 words somehow this novella reads more like a draft than a proper Season 9 mini book.
4. **New myth arc.** The “Frost Shepherd” gives the story a new existential threat different from the defeated Night King which I guess makes the plot line more interesting.

## Key Takeaways!
1. **PDF generation is quite easy.** One prompt + one code cell = downloadable mini-book in minutes. 
2. **Depth vs. breadth trade off.** Short-form prose can’t replicate eight hours of screen time, expect basic summary not heavy plot line.  
3. **Great start!** The draft functions as a storyboard: we can expand on the scenes and add more dialogue!
