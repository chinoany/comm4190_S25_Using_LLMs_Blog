# Sarcasm Detector Challenge: Can ChatGPT Spot the Snark or Take It Literally?

## Introduction
Sarcasm is highly context dependent with little to no literal signaling. Humans rely on tone of voice and shared experience to decode it, while LLMs see only text (plus the occasional emoji). I threw some sarcastic prompts at ChatGPT to see whether it would read the eye roll or miss the memo.


---

## Experiment Setup
- **Model:** GPT-4o  
- **Experiment:** Submit each sarcastic sentence exactly like someone would text, then ask the model to label the *intended meaning* (positive, neutral, negative) and explain why.

---

## Case 1  Obvious Verbal Irony  
> **First prompt:**  
> Label the sentiment and meaning of this sentence:  
> Great, another 7 a.m. meeting on Mondaymy favorite thing.  
>
> **GPT-4o chat response 1:**  
> **Sentiment**: Negative (sarcastic)
> **Meaning:** The speaker is frustrated or annoyed about having another early Monday meeting, despite sarcastically calling it their "favorite thing."
> Would you like a few rephrasings too?

**Observation:** Straightforward irony detected without hesitation.

---

## Case 2  Being Extra with the Sarcasm 
> **Second prompt:**  
> Label sentiment/meaning:  
> Oh perfect, my flight got delayed again. Love spending extra hours in the airport!  
>
> **GPT-4o chat response 2:**  
> **Sentiment:** Negative (sarcastic)
> **Meaning:** The speaker is frustrated and annoyed that their flight was delayed again, even though they sarcastically claim to enjoy waiting longer in the airport.
> Want me to show a few variations too?

**Observation:** Chat correctly interprets again!

---

## Case 3  Sarcasm Using a Common Joke
> **Third prompt:**  
> Label sentiment/meaning:  
> Yeah, because Windows updates never take long, right?  
>
> **GPT-4o chat response 3:**  
> **Sentiment:** Negative (sarcastic)
> **Meaning:** The speaker is criticizing how long Windows updates usually take, sarcastically pretending they are quick when they are actually slow.
> Would you like a few even sharper rewordings?

**Observation:** This interpretation required knowing the stereotype about Windows system being slow and Chat nailed it!

## Case 4  Emoji Masking Sarcasm  
> **Fourth prompt:**  
> Label sentiment/meaning:  
> So happy the internet broke during my livestream   
>
> **GPT-4o chat response 4:**  
> **Sentiment:** Negative  
> **Reasoning:** The smiling-teeth emoji conflicts with undesirable event (internet outage); mismatch signals sarcasm or forced cheer masking irritation.

**Observation:** Chat overrides the positive meaning of the emoji and points out hidden frustration.

---

## Case 5  Context Dependent  
> **Fifth prompt:**  
> Label sentiment/meaning:  
> Well, dont strain yourself helping, Todd.  
>
> **GPT-4o chat response 5:**  
> **Sentiment:** Neutral to mildly negative  
> **Reasoning:** Speaker might be cautioning Todd not to overexert himself.  
>
> **Reality:** Its a biting jab implying Todd is *not* helping. Model under-flags the sarcasm.

**Observation:** Without broader context or the tone used when saying the phrase, Chat offers two correct potential meanings.

---

## Key Takeaways!
1. **Cultural priors help.** The Windows-update joke showed ChatGPT understands tech lore.  
2. **Emoji clashes are handled.** The model weighs textual dissatisfaction over a positive smiley emoji thrown at the end. 
3. **Subtle digs still slip.** Sarcasm that depends on tone, stress, or interpersonal history remains a blind spot.

---

## Why Does This Matter?
Customer service bots, moderation tools, and sentiment dashboards cant afford to mistake sarcasm for praise (or vice-versa). Misclassifying sarcasm skews analytics and fuels tone deaf replies. Continuous testing can keeps your chatbot from taking every Great job  at face value.
