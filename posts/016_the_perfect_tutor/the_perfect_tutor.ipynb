Today I tested to see if ChatGPT could help me study for a quiz. I used my History of American Law Notes (a class I took my sophomore year) and had Chat come up with a way to teach me the material.

## Data Used  
A running 64-page document contaning notes on all class lectures and readings.

## Prompt Sequence  

1. **Outline request** – “Condense these notes into an outline”  
2. **Flashcards request** – “Create "flashcards"/identification terms and definitions covering the most testable details”  
3. **Practice quiz request** – “Write a multiple‑choice quiz with explanations”

## Key Takeaways  

1. **Time‑saver** – Ten minutes of prompting replaced hours of manual card creation  
2. **Gap finder** – Asking for “missing topics” surfaces blind spots fast
3. **Verify things** – AI sometimes hallucinates and adds fake topics; cross‑check with the online search and course syllabus
4. **Prompt tip** – End with “emphasize what’s likely on the quiz” to force prioritization.

---

## Why Does This Matter?  
For content‑heavy courses like American legal history, prepping can drown you in cases and terms/definitions. An LLM can triage the overload, summarizing, quizzing, and spotlighting gaps, so you spend brainpower on reasoning instead of re‑typing.
