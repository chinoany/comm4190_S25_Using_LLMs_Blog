# Lets Play Telephone!: Testing Google Gemini's ability to translate a phrase through multiple languages back to English

## Introduction  
Remember the playground game “telephone,” where a whispered sentence mutates as it passes down the line? Well today we'll be playing that but with a twist: multilingual and digital. I asked **Google's Gemini** to play a round of linguistic telephone:

> **English → German → Japanese → Binary (ASCII) → English**

Would the phrase survive intact, or come back garbled like playground gossip? Below I document the chat, check the responses, and note changes if any occur.

## Why Does This Matter?  
Algorithms are often used to translate (English → X → Y) due to resource gaps or cost. Each hop from language to language risks semantic drift, especially for idioms or less literal phrases that might not be able to directly translate to another language.
