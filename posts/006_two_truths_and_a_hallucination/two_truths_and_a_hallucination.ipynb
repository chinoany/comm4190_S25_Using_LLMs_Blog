# Two Truths & a Hallucination: Fact-Checking ChatGPT in Real Time

## Introduction
Large language models often sound so confident that even seasoned researchers will nod along, until you notice Antarctica has *tigers* and Mozart wrote “Bohemian Rhapsody.” To gauge how much scrutiny ChatGPT needs, I played “Two Truths & a Lie”: the model must supply **three** surprising facts on a topic, promising that *exactly one* is false. My job is to spot the hallucination and verify the truths!


---

## Experiment Setup
- **Model:** GPT-4o  
- **Experiment:**  
  1. Ask for three “surprising facts: two true, one false.”  
  2. Identify the claim that looks sus.  
  3. Fact check all three statements by searching them up on the internet. Reliable right?  
  4. Expose and note which fact was actually fake news.

---

## Case 1 – Space Exploration  
> **First prompt:** Give me three surprising facts about **Mars exploration**—two correct, one false but believable.  
> **GPT-4o chat response 1:**  
> 1. The **Curiosity** rover’s wheels bear Morse-code patterns that spell “JPL.”  
> 2. **Viking 1** carried a tiny piece of the Wright brothers’ 1903 Flyer.  
> 3. **Ingenuity** became the first aircraft to fly on another planet in 2021.  

**Fact-check:**  
1. True—wheel treads do encode “JPL.”  
2. **False**—Ingenuity carries Wright Flyer fabric; Viking 1 did not.  
3. True—Ingenuity’s 2021 flight is historic.

## Case 2 – Human Anatomy  
> **Second prompt:** Three surprising facts about **the human brain**—two true, one false.  
> **GPT-4o chat response 2:**  
> 1. The brain consumes roughly 20% of the body’s resting energy.  
> 2. Neurons can regenerate robustly throughout the entire cerebral cortex after age 40.  
> 3. The left hemisphere controls speech in about 90% of right-handed people.  

**Fact-check:**  
1. True—metabolic share is ~20%.  
2. **False**—adult cortical neurogenesis is minimal and still debated.  
3. True—classic laterality statistic.

## Case 3 – Pop-Culture Milestones  
> **Third prompt:** Three surprising facts about **the Oscars**—two true, one false.  
> **GPT-4o chat response 3:**  
> 1. Walt Disney once received four competitive Oscars in a single night.  
> 2. **Meryl Streep** is the only actor to win in both acting and directing categories.  
> 3. The statuette’s official name is the “Academy Award of Merit.”  

**Fact-check:**  
1. True—Disney won four in 1954.  
2. **False**—Streep has never won an Oscar for directing.  
3. True—“Academy Award of Merit” is the formal title.

---

## Key Takeaways!
1. **Hallucinations hide in details.** Dates, names, and “only ever” claims are usually where they fail/become false.  
2. **Prompting for a lie doesn’t stop lying.** The model happily inserts a falsehood and delivers it confidently.  
3. **Critical reading is mandatory.**  
4. **Pattern awareness helps.** Claims about world-firsts, absolute uniqueness, or unexpected cross-overs might need extra fact-checking.

---

## Why Does This Matter?
From classroom handouts to official research papers, unvetted AI “facts” can slip straight into print, eroding credibility and misinforming readers. Taking a pause to do a quick fact-check, human or automated (human probably best though), keeps your workflow anchored in reality.
