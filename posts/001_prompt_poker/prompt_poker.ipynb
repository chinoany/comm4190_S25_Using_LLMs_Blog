# Prompt Poker: How Small Wording Tweaks Change a Model’s Entire Hand

## Introduction
Change a single word in your prompt and ChatGPT reshuffles tone, depth, and even moral framing. Below I run three quick “hands,” keeping everything constant except one tiny phrase, then examine how the model’s output changes!


---

## Experiment Setup
- **Model:** GPT-4o  
- **Experiment:** Start with a baseline prompt, then alter only **one phrase** per round. All other parameters remain the same.

---

## Case 1 – Tone Shift  
*Changed word: **explain** -> **summarize***  

> **First prompt:** Explain climate change in **exactly 150 words**.  
> **GPT-4o chat response 1:**  
> Climate change refers to persistent shifts in Earth’s average temperature and weather regimes produced primarily by human-driven greenhouse-gas forcing. Burning coal, oil, and natural gas releases carbon dioxide, methane, and nitrous oxide, which absorb outgoing infrared radiation and re-emit it downward, reducing the planet’s net energy loss to space. As the lower troposphere warms, feedback loops magnify the effect: warmer air holds more water vapour, melting sea ice lowers albedo, and thawing permafrost liberates methane. The ocean absorbs most excess heat, leading to thermal expansion and sea-level rise, while altered temperature gradients reposition jet streams and storm tracks. Consequences include longer heatwaves, intensified drought-flood cycles, and ecosystem disruption. Ice-core records, isotopic signatures, and satellite radiometry together confirm the anthropogenic signal and bound equilibrium climate sensitivity near three degrees Celsius per CO₂ doubling.  
>
> **Second prompt:** Summarize climate change in **exactly 150 words**.  
> **GPT-4o chat response 2:**  
> Earth is warming faster than at any point in recorded history because people burn fossil fuels and cut down forests. Carbon-dioxide, methane, and other gases form a heat-trapping layer around the planet, acting like an extra blanket that holds in energy normally released to space. The extra warmth melts glaciers, pushes seas higher, and super-charges storms, droughts, and wildfires. Oceans soak up about 90 % of the added heat, expanding as they warm and turning more acidic, which endangers coral reefs and fisheries. Poorer communities and low-lying nations, although least responsible for emissions, feel the earliest, hardest impacts. Scientists agree—based on ice cores, tree rings, satellite data, and modern instruments—that the rise in temperature since the Industrial Revolution is human-driven and that limiting warming to 1.5 °C will require cutting global emissions roughly in half by 2030.

**Observation:** Both outputs meet the 150-word limit. However, *explain* gives a more mechanism heavy response (radiative forcing, feedbacks), whereas, *Summarize* uses general language to describe the consequences and moral framing.

## Case 2 – Audience Cue  
*Changed phrase: **“a 5-year-old”** -> **“my college roommate”***  
> **Third prompt:** Explain climate change to **a 5-year-old** in under 80 words.  
> **GPT-4o chat response 3:**  
> Climate change means Earth is getting hotter because people use things like cars and factories that make dirty air. The dirty air acts like a big blanket around the planet, trapping heat. This makes weather act weird — like stronger storms, really hot days, and melting ice where polar bears live. If we take better care of Earth by using clean energy and planting trees, we can help keep it safe and happy!
>
> **Fourth prompt:** Explain climate change to **my college roommate** in under 80 words.  
> **GPT-4o chat response 4:**  
> Climate change is the long-term heating of Earth’s climate system due to human activities, mainly burning fossil fuels like oil, gas, and coal. This traps heat in the atmosphere, leading to rising temperatures, stronger storms, droughts, and rising sea levels. It’s not just about the planet getting warmer — it’s about ecosystems, economies, and our future stability being at risk if we don’t shift to cleaner energy and adapt quickly.

**Observation:** Depending on the audience Chat edits its diction and metaphor use. The “5-year-old” respoonse uses blankets and feelings; the “roommate” version references sea-levels, emissions, and policy fixes (more appropriate for an older audience).

## Case 3 – Framing Bias  
*Changed set-up: **“What causes”** -> **“Who is to blame for”***  

> **Fifth prompt:** **What causes** climate change? Give two key factors.  
> **GPT-4o chat response 5:**  
> 1. **Greenhouse-gas emissions from fossil fuels**—Burning coal, oil, and natural gas releases CO₂, methane, and nitrous oxide that trap heat.  
> 2. **Land-use change**—Deforestation and industrial agriculture reduce carbon-sequestering vegetation and emit additional greenhouse gases.  
>
> **Sixth prompt:** **Who is to blame for** climate change? Give two key factors.  
> **GPT-4o chat response 6:**  
> 1. **High-income industrialized nations** whose historical and current fossil-fuel use account for the majority of cumulative greenhouse-gas emissions.  
> 2. **Large fossil-fuel corporations** that continue extracting, lobbying, and advertising despite decades-old evidence of climate risk.

**Observation:** Swapping *what* for *who* pushes the model from neutral causation toward moral attribution, naming specific actors and implying responsibility.

---

## Key Takeaways!
- **Prompt verbs steer depth.** *Explain* favors mechanism; *summarize* favors overview.  
- **Audience labels instantly retune register.** Explaining the reader/audience removes the need to tweak the output by hand.
- **Framing words can lead to bias.** Using *who* vs. *what* can lead to value judgments.
- **Small edits ≠ small effects.**

---

## Why Does This Matter?
If you treat prompts as insignificant, you may accidentally change the story your model tells, undermining brand voice, skewing analysis, or injecting unintended bias. Practicing “Prompt Poker” trains you to identify and deliberately use these hidden levers!
